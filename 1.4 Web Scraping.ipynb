# Importing Libraries and Driver

```python
import requests
import bs4
from bs4 import BeautifulSoup
import re
```

# Scraping and Parsing Wiki History and Politics Contents

```python
# Getting webpage content
headers = {'User-Agent': 'MGWikiBot/1.0 (MGsbbz@hotmail.com)'}
page = requests.get('https://en.wikipedia.org/wiki/Key_events_of_the_20th_century', headers=headers)
page.raise_for_status()
```

```python
# Parsing webpage content
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.title)
```

```python
# Obtaining main content, text and list items and storing in text object
content_div = soup.find('div', {'id': 'mw-content-text'})
texts = []
for tag in content_div.find_all(['p', 'li']):
    text = tag.get_text(' ', strip=True)
    if text:
        texts.append(text)

raw_text = ' '.join(texts)
raw_text[:500]
```

```python
# Initial cleaning of whitespace and citation markets
clean_text = re.sub(r'[\d+]', '', raw_text)
clean_text = re.sub(r'\s+', " ", clean_text).strip()
clean_text[:500]
```

# Saving Contents as Text File

```python
# Saving the file to the working directory
with open('20th_century_article.txt', 'w', encoding='utf-8') as f:
    f.write(str(clean_text))
```
